```python
%load_ext autoreload
%autoreload 2

import os

import numpy as np
import h5py
import matplotlib.pyplot as plt
import unyt as u
import pandas as pd

import richio
richio.plots.use_nice_style()
```

```python
data_n1e3 = pd.read_csv('../scripts/N1e3_dissipation_RICH_vs_Tds.csv')
data_n1e4 = pd.read_csv('../scripts/N1e4_dissipation_RICH_vs_Tds.csv')

data_n1e3.sort_values(by='Time', inplace=True)
data_n1e4.sort_values(by='Time', inplace=True)
```

```python
acc_n1e3 = data_n1e3['RICH Dissipation'] / data_n1e3['Tds Dissipation']
acc_n1e4 = data_n1e4['RICH Dissipation'] / data_n1e4['Tds Dissipation']
plt.plot(data_n1e3['Time'], acc_n1e3, marker='.', label='Np=1e3')
plt.plot(data_n1e4['Time'][10:], acc_n1e4[10:], marker='.', label='Np=1e4')
plt.xlabel('Time')
plt.ylabel('Diss(RICH) / Diss(TdS)')
plt.axhline(1, color='k')
plt.ylim(0, 2)
plt.legend()
plt.show()
```
